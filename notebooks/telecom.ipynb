# Data Prep

```python
import sys
from IPython.display import display
sys.path.append('..')
from scripts.db import returnFromDb

#data = pd.read_csv('../docs/week2_data.csv')
data = returnFromDb()
display("head: ",data.head())
display("info: \n",data.info())
display("description stat: \n",data.describe())
```

```python
print(data.columns)
```

```python
# Count top 10 handsets used by customers
from decimal import Decimal
top_10_handsets = data['IMEI'].value_counts().head(10)
print(top_10_handsets)
```

#### Top 10 handsets

```python
top_10_handsets.plot(kind='barh', color='skyblue')
plt.title('Top 10 Handsets Used by Customers')
plt.xlabel('Count')
plt.ylabel('Handset')
plt.show()
```

```python
# Identify the top 3 handset manufacturers
top_3_manufacturers = data['Handset Manufacturer'].value_counts().head(3)

# Plot the top 3 handset manufacturers
top_3_manufacturers.plot(kind='bar', color='lightgreen')
plt.title('Top 3 Handset Manufacturers')
plt.xlabel('Manufacturers')
plt.ylabel('Count')
plt.show()
```

```python
# Loop through the top 3 manufacturers and find the top 5 handsets for each
for manufacturer in top_3_manufacturers.index:
    print(f"Top 5 handsets for {manufacturer}:")
    top_5_handsets = data[data['Handset Manufacturer'] == manufacturer]['IMEI'].value_counts().head(5)
    print(top_5_handsets)
    print("\n")
```

```python
# Data Aggregation
# Aggregating session data per user
user_behavior = data.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',    # Number of xDR sessions
    'Dur. (ms)': 'sum',      # Session duration
    'Avg Bearer TP DL (kbps)': 'sum',  # Total download data
    'Avg Bearer TP UL (kbps)': 'sum',  # Total upload data
}).rename(columns={
    'Bearer Id': 'xDR_sessions', 
    'Dur. (ms)': 'Total_duration', 
    'Avg Bearer TP DL (kbps)': 'Total_DL', 
    'Avg Bearer TP UL (kbps)': 'Total_UL'
})

# Add total data volume column
user_behavior['Total_data_volume'] = user_behavior['Total_DL'] + user_behavior['Total_UL']
print(user_behavior.head())
```
