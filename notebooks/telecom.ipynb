```python
def treat_outliers_all_columns(df):
    for col in df.select_dtypes(include=['number']).columns:
        # Calculate the 95th percentile
        q95 = np.quantile(df[col], 0.95)

        # Identify outliers
        outliers = df[col] > q95

        # Replace outliers with the median
        df.loc[outliers, col] = df[col].median()

    return df

treat_outliers_all_columns(data)
```

```python
def plot_dispersion_side_by_side(dispersion_summary, columns_per_row=2):
    """
    Creates side-by-side plots for dispersion analysis.

    Args:
        dispersion_summary (pd.DataFrame): DataFrame containing dispersion metrics.
        columns_per_row (int, optional): Number of columns per row in the plot. Defaults to 2.
    """

    num_columns = len(dispersion_summary.columns)
    num_rows = int(np.ceil(num_columns / columns_per_row))  # Calculate number of rows

    fig, axes = plt.subplots(num_rows, columns_per_row, figsize=(10 * columns_per_row, 6 * num_rows))
    axes = axes.ravel()  # Flatten the axes for easier iteration

    for i, column in enumerate(dispersion_summary.columns):
        metric_types = dispersion_summary.index

        for j, metric in enumerate(metric_types):
            axes[i * columns_per_row + j].plot(dispersion_summary[column])
            axes[i * columns_per_row + j].set_title(f"{metric} - {column}")
            axes[i * columns_per_row + j].set_xlabel(column)

        if i == num_rows - 1:  # Adjust x-axis labels only for the bottom row
            for ax in axes[i * columns_per_row:]:
                ax.set_xlabel(column)

        axes[i * columns_per_row].set_ylabel(metric_types[0])  # Set y-label for the first plot in each row

    fig.suptitle('Dispersion Analysis - Side-by-Side Plots', fontsize=14)
    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()

plot_dispersion_side_by_side(dispersion_parameters, columns_per_row=3)  # Adjust columns_per_row as needed
```

```python
# Aggregate metrics for each user (sessions, duration, traffic)
engagement_metrics = data.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',    # Number of sessions
    'Dur. (ms)': 'sum',      # Session duration
    'Avg Bearer TP DL (kbps)': 'sum',  # Download traffic
    'Avg Bearer TP UL (kbps)': 'sum',  # Upload traffic
})

# Normalize the engagement metrics
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
normalized_metrics = scaler.fit_transform(engagement_metrics)

# Perform k-means clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0)
engagement_metrics['cluster'] = kmeans.fit_predict(normalized_metrics)
```

```python
# Use elbow method to determine optimal k
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(normalized_metrics)
    wcss.append(kmeans.inertia_)

# Plot elbow method
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
