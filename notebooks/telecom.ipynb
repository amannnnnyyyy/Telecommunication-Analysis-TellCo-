## Data Cleaning

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)


# Fill missing categorical columns with mode (most frequent value)
for col in data.select_dtypes(include=['object']).columns:
    data[col].fillna(data[col].mode()[0], inplace=True)

#numerical
for col in data.select_dtypes(include=[int, float]).columns:
    # Check if there are any non-null values in the column
    if data[col].notnull().any():
        data[col].fillna(data[col].mean(), inplace=True)
```

```python
#Recheck
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Ensure all columns have the correct data type
data.dtypes
```

```python
def treat_outliers_all_columns(df):
    for col in df.select_dtypes(include=['number']).columns:
        # Calculate the 95th percentile
        q95 = np.quantile(df[col], 0.95)

        # Identify outliers
        outliers = df[col] > q95

        # Replace outliers with the median
        df.loc[outliers, col] = df[col].median()

    return df

treat_outliers_all_columns(data)
```
